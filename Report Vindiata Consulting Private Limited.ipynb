```python
# Rename columns for clarity
df_gameplay.rename(columns={'User ID': 'User_ID', 'Games Played': 'Games_Played', 'Datetime': 'Timestamp'}, inplace=True)
df_deposit.rename(columns={'User Id': 'User_ID', 'Datetime': 'Timestamp', 'Amount': 'Deposit_Amount'}, inplace=True)
df_withdrawal.rename(columns={'User Id': 'User_ID', 'Datetime': 'Timestamp', 'Amount': 'Withdrawal_Amount'}, inplace=True)
```

```python
print("User Gameplay data.csv - Cleaned Head:")
print(df_gameplay.head())
print("\nUser Gameplay data.csv - Cleaned Info:")
print(df_gameplay.info())

print("\nDeposit Data.csv - Cleaned Head:")
print(df_deposit.head())
print("\nDeposit Data.csv - Cleaned Info:")
print(df_deposit.info())

print("\nWithdrawal Data.csv - Cleaned Head:")
print(df_withdrawal.head())
print("\nWithdrawal Data.csv - Cleaned Info:")
print(df_withdrawal.info())
```

```python
# Convert 'Timestamp' columns to datetime objects
df_gameplay['Timestamp'] = pd.to_datetime(df_gameplay['Timestamp'], format='%d-%m-%Y %H:%M')
df_deposit['Timestamp'] = pd.to_datetime(df_deposit['Timestamp'], format='%d-%m-%Y %H:%M')
df_withdrawal['Timestamp'] = pd.to_datetime(df_withdrawal['Timestamp'], format='%d-%m-%Y %H:%M')
```

```python
# Extract Date and Hour for Slot calculation
df_gameplay['Date'] = df_gameplay['Timestamp'].dt.date
df_gameplay['Hour'] = df_gameplay['Timestamp'].dt.hour
df_deposit['Date'] = df_deposit['Timestamp'].dt.date
df_deposit['Hour'] = df_deposit['Timestamp'].dt.hour
df_withdrawal['Date'] = df_withdrawal['Timestamp'].dt.date
df_withdrawal['Hour'] = df_withdrawal['Timestamp'].dt.hour
```

```python
# Determine Slot (S1: 12am-12pm, S2: 12pm-12am)
def get_slot(hour):
    return 'S1' if hour < 12 else 'S2'

df_gameplay['Slot'] = df_gameplay['Hour'].apply(get_slot)
df_deposit['Slot'] = df_deposit['Hour'].apply(get_slot)
df_withdrawal['Slot'] = df_withdrawal['Hour'].apply(get_slot)
```

```python
# Convert Date to datetime for easier filtering
df_gameplay['Date'] = pd.to_datetime(df_gameplay['Date'])
df_deposit['Date'] = pd.to_datetime(df_deposit['Date'])
df_withdrawal['Date'] = pd.to_datetime(df_withdrawal['Date'])

print("\nDataFrames after adding Date and Slot columns - Head:")
print(df_gameplay.head())
print(df_deposit.head())
print(df_withdrawal.head())
```

```python
# --- Loyalty Point Calculation Function ---
def calculate_loyalty_points(df_agg):
    """
    Calculates loyalty points based on aggregated player data.
    df_agg should have columns: 'User_ID', 'Total_Deposits', 'Total_Withdrawals',
                               'Num_Deposits', 'Num_Withdrawals', 'Total_Games_Played'
    """
    # Initialize Loyalty Points column
    df_agg['Loyalty_Points'] = 0.0

    # (0.01 * deposit)
    df_agg['Loyalty_Points'] += 0.01 * df_agg['Total_Deposits']

    # (0.005 * Withdrawal amount)
    df_agg['Loyalty_Points'] += 0.005 * df_agg['Total_Withdrawals']

    # (0.001 * maximum of (#deposit - #withdrawal) or 0)
    df_agg['deposit_withdrawal_diff'] = df_agg['Num_Deposits'] - df_agg['Num_Withdrawals']
    df_agg['Loyalty_Points'] += 0.001 * df_agg['deposit_withdrawal_diff'].apply(lambda x: max(x, 0))

    # (0.2 * Number of games played)
    df_agg['Loyalty_Points'] += 0.2 * df_agg['Total_Games_Played']

    return df_agg

# --- Part A, Question 1: Slot-wise Loyalty Points ---
print("\n--- Part A, Question 1: Slot-wise Loyalty Points ---")

slots_to_calculate = {
    '2nd October Slot S2': {'date': '2022-10-02', 'slot': 'S2'},
    '6th October Slot S2': {'date': '2022-10-06', 'slot': 'S2'},
    '8th October Slot S1': {'date': '2022-10-08', 'slot': 'S1'},
    '26th October Slot S2': {'date': '2022-10-26', 'slot': 'S2'},
}

loyalty_points_slotwise = {}

for desc, params in slots_to_calculate.items():
    target_date = pd.to_datetime(params['date']).date()
    target_slot = params['slot']

    # Filter data for the specific date and slot
    gameplay_filtered = df_gameplay[(df_gameplay['Date'].dt.date == target_date) & (df_gameplay['Slot'] == target_slot)]
    deposit_filtered = df_deposit[(df_deposit['Date'].dt.date == target_date) & (df_deposit['Slot'] == target_slot)]
    withdrawal_filtered = df_withdrawal[(df_withdrawal['Date'].dt.date == target_date) & (df_withdrawal['Slot'] == target_slot)]

    # Aggregate data for loyalty point calculation
    agg_gameplay = gameplay_filtered.groupby('User_ID').agg(
        Total_Games_Played=('Games_Played', 'sum')
    ).reset_index()

    agg_deposit = deposit_filtered.groupby('User_ID').agg(
        Total_Deposits=('Deposit_Amount', 'sum'),
        Num_Deposits=('User_ID', 'count')
    ).reset_index()

    agg_withdrawal = withdrawal_filtered.groupby('User_ID').agg(
        Total_Withdrawals=('Withdrawal_Amount', 'sum'),
        Num_Withdrawals=('User_ID', 'count')
    ).reset_index()

    # Merge aggregated dataframes
    df_merged_slot = pd.merge(agg_gameplay, agg_deposit, on='User_ID', how='outer')
    df_merged_slot = pd.merge(df_merged_slot, agg_withdrawal, on='User_ID', how='outer')

    # Fill NaN values with 0 for calculations (e.g., if a player didn't deposit in a slot)
    df_merged_slot.fillna(0, inplace=True)

    # Calculate loyalty points
    df_loyalty_slot = calculate_loyalty_points(df_merged_slot.copy()) # Use .copy() to avoid SettingWithCopyWarning
    loyalty_points_slotwise[desc] = df_loyalty_slot[['User_ID', 'Loyalty_Points', 'Total_Games_Played', 'Total_Deposits', 'Total_Withdrawals', 'Num_Deposits', 'Num_Withdrawals']].sort_values(by='Loyalty_Points', ascending=False)
    print(f"\n--- {desc} ---")
    print(loyalty_points_slotwise[desc].head())
    loyalty_points_slotwise[desc].to_csv(f'Loyalty_Points_{desc.replace(" ", "_")}.csv', index=False)
```

```python
# --- Part A, Question 2: Overall Loyalty Points and Rank for October ---
print("\n--- Part A, Question 2: Overall Loyalty Points and Rank for October ---")

# Aggregate data for the entire month of October
# Note: The 'Date' columns are already datetime objects and include all of October 2022.

agg_gameplay_overall = df_gameplay.groupby('User_ID').agg(
    Total_Games_Played=('Games_Played', 'sum')
).reset_index()

agg_deposit_overall = df_deposit.groupby('User_ID').agg(
    Total_Deposits=('Deposit_Amount', 'sum'),
    Num_Deposits=('User_ID', 'count')
).reset_index()

agg_withdrawal_overall = df_withdrawal.groupby('User_ID').agg(
    Total_Withdrawals=('Withdrawal_Amount', 'sum'),
    Num_Withdrawals=('User_ID', 'count')
).reset_index()

# Merge aggregated dataframes for overall calculation
df_merged_overall = pd.merge(agg_gameplay_overall, agg_deposit_overall, on='User_ID', how='outer')
df_merged_overall = pd.merge(df_merged_overall, agg_withdrawal_overall, on='User_ID', how='outer')

# Fill NaN values with 0
df_merged_overall.fillna(0, inplace=True)

# Calculate overall loyalty points
df_loyalty_overall = calculate_loyalty_points(df_merged_overall.copy())

# Rank players
df_loyalty_overall['Rank'] = df_loyalty_overall.sort_values(
    by=['Loyalty_Points', 'Total_Games_Played'],
    ascending=[False, False]
).reset_index().index + 1

print("\nOverall Loyalty Points and Rank (Top 10):")
print(df_loyalty_overall[['User_ID', 'Loyalty_Points', 'Total_Games_Played', 'Rank']].sort_values(by='Rank').head(10))

# Save overall loyalty points to CSV
df_loyalty_overall.to_csv('Overall_Loyalty_Points_October.csv', index=False)


# --- Part A, Question 3: Average Deposit Amount ---
print("\n--- Part A, Question 3: Average Deposit Amount ---")

average_deposit_amount = df_deposit['Deposit_Amount'].mean()
print(f"Average Deposit Amount: ${average_deposit_amount:,.2f}")

# --- Part A, Question 4: Average Deposit Amount Per User ---
print("\n--- Part A, Question 4: Average Deposit Amount Per User ---")

total_deposit_per_user = df_deposit.groupby('User_ID')['Deposit_Amount'].sum().reset_index()
average_deposit_per_user = total_deposit_per_user['Deposit_Amount'].mean()
print(f"Average Deposit Amount Per User: ${average_deposit_per_user:,.2f}")

# --- Part A, Question 5: Average Number of Games Played Per User ---
print("\n--- Part A, Question 5: Average Number of Games Played Per User ---")

total_games_per_user = df_gameplay.groupby('User_ID')['Games_Played'].sum().reset_index()
average_games_per_user = total_games_per_user['Games_Played'].mean()
print(f"Average Number of Games Played Per User: {average_games_per_user:,.2f}")
```

```python
Part C
Would you say the loyalty point formula is fair or unfair?
The loyalty point formula (Loyalty Point = (0.01 * deposit) + (0.005 * Withdrawal amount) + (0.001 * maximum of (#deposit - #withdrawal) or 0) + (0.2 * Number of games played))  is generally fair, as it considers multiple aspects of player engagement: deposits, withdrawals, and game participation. It appropriately assigns higher weight to deposits, which directly contribute to company revenue, and incentivizes active gameplay.



However, there's a potential point of ambiguity or perceived unfairness in the component 0.001 * maximum of (#deposit - #withdrawal) or 0. This part of the formula only awards points if the number of deposits exceeds the number of withdrawals, giving zero points otherwise. This could be seen as unfair to consistently winning players who actively engage in games and generate platform fees but make more withdrawals than deposits. It prioritizes deposit frequency over overall gameplay and win/loss activity.
```

```python
Can you suggest any way to make the loyalty point formula more robust?
To make the loyalty point formula more robust and comprehensive, consider these suggestions:

Refine Deposit vs. Withdrawal Component:

Net Deposit Amount: Instead of just counts, consider incorporating a "Net Deposit Amount" (Total Deposits - Total Withdrawals). This would reward players who bring more capital to the platform, irrespective of transaction frequency.
Positive Points for All Activities: Assign points for both deposits and withdrawals, but with different weightages, ensuring that all forms of active engagement are acknowledged.
Incorporate Time Spent Playing: Beyond just the number of games played, tracking the actual duration of active gameplay could be a stronger indicator of engagement and loyalty. This could be given a smaller, incremental point value.

Reward for Consistency/Streaks: Introduce bonus points for consistent daily logins, weekly game streaks, or continuous engagement over a period. This fosters routine usage of the platform.

Tiered Rewards for High-Value Actions: For large deposits or significant game milestones, consider implementing tiered or accelerated point accumulation. For example, a deposit of $10,000 might yield slightly more than proportional points compared to ten $1,000 deposits, encouraging higher value transactions.

Incentivize Referrals: If player acquisition is a goal, award loyalty points for successfully referring new users who become active players.

Dynamic Weightage Adjustment: Regularly review and adjust the weightages of different loyalty point components based on strategic business objectives (e.g., if the goal is to increase game engagement, increase the weight for "Number of games played").
```
